```

history={}
history['fold1']=np.load('history_all_fold_1_Resnet_rmsprop.npy',allow_pickle=True).item()
history['fold2']=np.load('history_all_fold_2_Resnet_rmsprop.npy',allow_pickle=True).item()
history['fold3']=np.load('history_all_fold_3_Resnet_rmsprop.npy',allow_pickle=True).item()
history['fold4']=np.load('history_all_fold_4_Resnet_rmsprop.npy',allow_pickle=True).item()
history['fold5']=np.load('history_all_fold_5_Resnet_rmsprop.npy',allow_pickle=True).item()

```

```
from matplotlib import pyplot as plt
for i in range(5):
    fold='fold'+str(i+1)
    fold1='fold_'+str(i+1)
    plt.plot(history[fold][fold1]['accuracy'])
    plt.title('loss for fold '+str(i))
    plt.xlabel('epoch')
    plt.ylabel('accuracy')
    plt.title('Training all layers')
    plt.show()
```
